Figure 8.2: The net flux through an area A

```python
fig = plt.figure(figsize = (4, 2))
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)

for direction in ["xzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)


# removing the axis ticks
plt.xticks([]) # labels 
plt.yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

# Area, A
ax.plot(np.zeros(20)+4.5, np.linspace(0.3, 2.2, 20), color = 'indianred', lw = 4)
ax.annotate('A', (4.6, 2), (6, 2.3), arrowprops=({'arrowstyle': '->', 'lw': 1, 
            'color':'indianred'}), va = 'center', color = 'indianred', fontsize = 8)

# Number of molecules to the left of x, N(x)
ax.text(2, 0.1, '$N(x)$', fontsize = 9)
# Half of the N(x) molecules moving past x
ax.annotate(r'$\frac{N(x)}{2}$', (5, 1.5), (1, 1.5), arrowprops=({'arrowstyle': '->', 'lw': 1, 
            'facecolor':'k'}), va = 'center', fontsize = 12)
# Number of molecules to the right of x, N(x+dx)
ax.text(6, 0.1, '$N(x+dx)$', fontsize = 9)
# Half of the N(x+dx) molecules moving past x
ax.annotate(r'$\frac{N(x+dx)}{2}$', (4, 1), (8, 1), arrowprops=({'arrowstyle': '->', 'lw': 1, 
            'facecolor':'k'}), va = 'center', fontsize = 12)

# Ax tick label, x
ax.scatter(4.5, 0, color = 'k', marker = 2)
ax.text(4.5, -0.2, '$x$', ha = 'center', fontsize = 9)

ax.set_xlim(0, 10)
ax.set_ylim(0, 2.5)

plt.tight_layout()
plt.savefig('Flux.png', dpi = 150)
plt.show()
```

Figure 8.4: A Brownian particle in a solution

```python
# Vector field showing the solution
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = -y/np.sqrt(x**2 + y**2)
v = x/np.sqrt(x**2 + y**2)

# Source: https://krajit.github.io/sympy/vectorFields/vectorFields.html
```

```python
fig, ax = plt.subplots(1,1, figsize = (5, 3))

# Vector field showing the solution
ax.quiver(x,y,u,v, headwidth = 0.1, width = 0.002)
# Brownian particle
ax.scatter(0, 0, color = 'steelblue', s = 400)
ax.text(-0.45, 0, 'm', color = 'steelblue', va = 'center')
ax.text(0, -0.8, '$x(t)$', ha = 'center')
# Velocity vector of particle
ax.quiver(0, 0, 3, 2, width = 0.004, color = 'darkred', angles='xy', scale_units='xy', scale=4.5)
ax.text(0.4, -0.1, '$v(t)$', color = 'darkred')

plt.xticks([], [])
plt.yticks([], [])
ax.set_xticks([]) # for major ticks
ax.set_yticks([])
ax.set_xticks([], minor=True) # for minor ticks
ax.set_yticks([], minor=True)
# sns.despine(top=True, right=True, left=True, bottom=True)

ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)

plt.tight_layout()
plt.savefig('Brown.png', dpi = 150)
plt.show()
```

Figure 8.5: Concentration profiles as a function of time and distance from the surface of the electrode

```python
eta = 10**(-3) # Viscosity of solvent in Pa*s
a = 5*10**(-11) #Radius of particle in m
k = 1.381*10**(-23) # Boltzmann konstant in J/K
T = 295 # Temperature in K

alpha = 6*np.pi*eta*a # Friction coefficient
D = (k*T)/alpha # Diffusion coefficient

# Distance from the surface of the electrode
x = np.linspace(0, 4*10**(-5), 1000)

# Time
t_1 = 0.001
t_2 = 0.01
t_3 = 0.1
t_4 = 1

# Concentration profiles
C_1 = special.erf(x/np.sqrt(2*D*t_1))
C_2 = special.erf(x/np.sqrt(2*D*t_2))
C_3 = special.erf(x/np.sqrt(2*D*t_3))
C_4 = special.erf(x/np.sqrt(2*D*t_4))
```

```python
fig = plt.figure(figsize = (5, 3))
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)


# removing the axis ticks
plt.xticks([]) # labels 
plt.yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

# Concentration profile at t = 0.001s
ax.plot(x, C_1, lw = 1.1, color = 'tab:blue')
ax.text(x[100], 0.9, '$t=0.001$s', color = 'tab:blue', fontsize = 9)
# Concentration profile at t = 0.01s
ax.plot(x, C_2, lw = 1.1, color = 'tab:orange')
ax.text(x[210], 0.75, '$t=0.01$s', color = 'tab:orange', fontsize = 9)
# Concentration profile at t = 0.1s
ax.plot(x, C_3, lw = 1.1, color = 'tab:green')
ax.text(x[430], 0.53, '$t=0.1$s', color = 'tab:green', fontsize = 9)
ax.plot(x, C_4, lw = 1.1, color = 'tab:red')
# Concentration profile at t = 1s
ax.text(x[600], 0.22, '$t=1$s', color = 'tab:red', fontsize = 9)
# Concentration profile at t = 0s
ax.hlines(1, 0, 4*10**(-5), color = 'k', lw = 1.1)
ax.text(x[550], 1.02, '$t=0$', fontsize = 9)
ax.text (-x[25], 1, '1', va = 'center')

ax.set_ylim(0,1.2)
ax.set_xlim(0, 4*10**(-5))

ax.set_ylabel(r'$\frac{C_{P}}{C_{P}^{*}}$')
ax.axis['yzero'].label.set(rotation = 'vertical', fontsize = 12)
ax.set_xlabel('x')
ax.axis['xzero'].label.set(fontsize = 12)

plt.tight_layout()
plt.savefig('CProf.png', dpi = 150)
plt.show()
```
