```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.axisartist.axislines import SubplotZero
import matplotlib.patches as mpatches
import matplotlib
```

Figure 12.1: The potential U(x)

```python
x = np.linspace(-5, 10, 1000)
y = x**3-8*x**2+14*x
```

```python
fig = plt.figure(figsize = (5,4))
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)


# removing the axis ticks
plt.xticks([]) # labels 
plt.yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)

ax.plot(x[250:775]+3,y[250:775][::-1]+40, color = 'k')
ax.vlines(4.05, 0, 32, color = 'k', ls = '--', lw = 0.9, zorder = -10)
ax.vlines(7.3, 0, 47, color = 'k', ls = '--', lw = 0.9, zorder = -10)
ax.hlines(31.5, 4.05, 7.3, color = 'k', ls = '--', lw = 0.9, zorder = -10)

ax.text(6.7, 35, r'E$_{b}$')
ax.text(3.3, 30, r'$\omega_{0}$')
ax.text(4.05, -4, r'x$_{a}$', ha = 'center')
ax.text(7.3, -4, r'x$_{b}$', ha = 'center')

ax.set_ylim(0, 80)
ax.set_xlim(0, 12)

ax.set_ylabel('U(x)')
ax.set_xlabel(' \n Reaction coordinate, x')

fig.tight_layout()
plt.savefig('Kramer.png', dpi = 150)
plt.show()
```

Figure 12.2: The bistable potential as a function of the reaction coordinate

```python
RC = np.linspace(-7, 7, 1000)

y = 7*RC+4.9*RC**2-3.73*RC**3+0.5*RC**4
```

```python
fig = plt.figure(figsize = (6,4))
ax = SubplotZero(fig, 111)
fig.add_subplot(ax)

for direction in ["xzero", "yzero"]:
    # adds arrows at the ends of each axis
    ax.axis[direction].set_axisline_style("-|>")

    # adds X and Y-axis from the origin
    ax.axis[direction].set_visible(True)


# removing the axis ticks
plt.xticks([]) # labels 
plt.yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')

for direction in ["left", "right", "bottom", "top"]:
    # hides borders
    ax.axis[direction].set_visible(False)


ax.plot(RC+3, y+15, color = 'k')
ax.vlines(7.23, 10, 27, color = 'k', ls = '--', lw = 0.9)
ax.hlines(27, 1.58, 8.25, color = 'k', ls = '--', lw = 0.9, zorder = -10)
ax.annotate(' ', xy = (2.55, 12.9), xytext = (2.55, 27.9), arrowprops=({'arrowstyle': '<->', 'lw': 0.9, 'color':'k'}))

ax.text(2.65, 19.5, r'E$_{b}^{+}$')
ax.text(2.42, 11.2, r'$\omega_{a}$')
ax.text(6.8, 19.5, r'E$_{b}^{-}$')
ax.text(7.11, 8.2, r'$\omega_{c}$')

ax.scatter(2.55, 0, marker = 2, color ='k')
ax.scatter(4.8, 0, marker = 2, color = 'k')
ax.scatter(7.23, 0, marker = 2, color = 'k')
ax.text(2.55, -2, r'x$_{a}$', ha = 'center')
ax.text(4.8, -2, r'x$_{b}$', ha = 'center')
ax.text(7.23, -2, r'x$_{c}$', ha = 'center')

arrow = mpatches.FancyArrow(3.8, 29, 2, 0, length_includes_head = True, head_width = 0.5, head_length = 0.09, color = 'k')
ax.add_patch(arrow)
arrow = mpatches.FancyArrow(5.8, 32.5, -2, 0, length_includes_head = True, head_width = 0.5, head_length = 0.09, color = 'k')
ax.add_patch(arrow)
ax.text(4.9, 29.5, r'k$^{+}$', ha = 'center')
ax.text(4.9, 33, r'k$^{-}$', ha = 'center')

ax.set_ylim(0, 40)
ax.set_xlim(0, 10)

ax.set_xlabel('\n Reaction coordinate, x')
ax.set_ylabel('U(x)')

fig.tight_layout()
plt.savefig('Bistable.png', dpi = 150)
plt.show()
```
